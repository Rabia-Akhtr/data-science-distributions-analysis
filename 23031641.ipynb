```python
import pandas as pd
import matplotlib.pyplot as plt

def calculate_mean(data):
    """Calculate the mean of a list of numbers."""
    return sum(data) / len(data) if data else 0

def calculate_std(data):
    """Calculate the standard deviation of a list of numbers."""
    mean = calculate_mean(data)
    variance = sum((x - mean) ** 2 for x in data) / len(data) if data else 0
    return variance ** 0.5

def calculate_v_value(data):
    """Calculate the V value (grade such that 10% of students scored higher) for the dataset."""
    sorted_data = sorted(data)
    index_V = int(0.9 * len(sorted_data))  # 90% of students scored lower than this index
    return sorted_data[index_V]

def process_data_2020(data_2020):
    """Process 2020 data for histogram and calculations."""
    x_2020 = []
    for _, row in data_2020.iterrows():
        interval_mean = (row[0] + row[1]) / 2
        count = row[2]
        x_2020.extend([interval_mean] * count)
    return x_2020

# Correctly reading the datasets (replace with actual paths)
file_2020 = '2020input1.csv'
file_2024 = '2024input1.csv'

data_2020_corrected = pd.read_csv(file_2020, delim_whitespace=True)
data_2024_corrected = pd.read_csv(file_2024, delim_whitespace=True)

# Processing the 2020 dataset
x_2020 = process_data_2020(data_2020_corrected)

# Manual calculations of mean and standard deviation for 2020 and 2024 data
mean_2020 = calculate_mean(x_2020)
std_2020 = calculate_std(x_2020)
mean_2024 = calculate_mean(data_2024_corrected['64'].tolist())
std_2024 = calculate_std(data_2024_corrected['64'].tolist())

# Calculating the V value for 2024 data
V_value = calculate_v_value(data_2024_corrected['64'].tolist())

# Student ID (replace with actual ID)
student_id = "633"

# Creating the plot with specified colors
plt.figure(figsize=(10, 6))
plt.hist(x_2020, bins=range(0, 101, 5), color='blue', alpha=0.7, label='2020 Grades')
plt.hist(data_2024_corrected['64'], bins=range(0, 101, 5), color='orange', alpha=0.6, label='2024 Grades')

# Adding mean, std, V value, and other details in a text box
textstr = '\n'.join((
    f'Student ID: {student_id}',
    f'Mean 2020: {mean_2020:.2g}',
    f'STD 2020: {std_2020:.2g}',
    f'Mean 2024: {mean_2024:.2g}',
    f'STD 2024: {std_2024:.2g}',
    f'V value: {V_value}'
))
props = dict(boxstyle='round', facecolor='white', alpha=0.5)
plt.text(0.05, 0.95, textstr, transform=plt.gca().transAxes, fontsize=14,
        verticalalignment='top', bbox=props)

# Plotting details
plt.xlabel('Grades')
plt.ylabel('Number of Students')
plt.title('Grade Distributions for 2020 and 2024')
plt.legend()
plt.show()
```
